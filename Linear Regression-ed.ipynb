```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = [(1,8), (2,8), (3,6), (4,6), (5,4)]
```

```python
points
```

```python
xs = [x for x,y in points]
ys = [y for x,y in points]
```

```python
plt.plot(xs, ys, 'o')
```

```python
x_mean = np.mean(xs)
y_mean = np.mean(ys)

x_mean, y_mean
```

```python
inside = np.sum([(x - x_mean)**2 for x, y in points]) *  np.sum([(y - y_mean)**2 for x, y in points]) 
nominator = np.sum([(x - x_mean) * (y - y_mean) for x,y in points])
denominator = np.sqrt(inside)

r = nominator/denominator


(inside, denominator), r
```

```python
b_1 = np.sum([(x-x_mean) * (y-y_mean) for x,y in points]) / np.sum([(x-x_mean)**2 for x,y in points])
b_0 = y_mean - b_1 * x_mean

b_1, b_0
```

```python
p = plt.plot(xs, ys, 'o', xs, b_0 + b_1 * np.array(xs))

```
